```python
df.drop(columns=['id','size_employees','funding_stage','revenue_bucket','application_link','interview_stages','last_updated','company_rating'],inplace = True)
```

```python
df.head()
```

```python
text_columns = ['required_skills', 'preferred_degree', 'remote_policy', 'sector', 'company_type', 'hq_city', 'hq_state']

df['tags'] = df[text_columns].apply(
    lambda row: ' '.join(
        str(item).lower().replace(" ", "")   # lowercase + remove spaces inside words
        for item in row if pd.notnull(item)  # skip NaN values
    ),
    axis=1
)
```

```python
df['tags'][0]

new_df = df[['company_name','internship_openings','stipend_min_inr','stipend_max_inr','cgpa_cutoff','tags']]

new_df.head()

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    lowercase=True,
    stop_words=None,
    token_pattern=r'(?u)\b\w+\b',
    ngram_range=(1, 2),
    max_df=1.0,
    min_df=1,
    max_features=5000,
    binary=True
)
```

```python
# Keep as sparse matrix (recommended for large datasets)
vectors = cv.fit_transform(new_df['tags'])

print("Shape of vectorized matrix:", vectors.shape)
print("Sample feature names:", cv.get_feature_names_out()[:20])

# If you really need a dense array (be careful with memory usage)
dense_vectors = vectors.toarray()


dense_vectors
```

```python
len(cv.get_feature_names_out())
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem_text(text):
    return " ".join(ps.stem(word) for word in text.split())

# Example: apply to your tags column
new_df['tags'] = new_df['tags'].apply(stem_text)


new_df.head()
```

```python
new_df['tags'][0]

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Function to compute recommendations using chunked cosine similarity
def recommend_internships(index, top_n=5, batch_size=1000):
    n = vectors.shape[0]
    sim_scores = np.zeros(n, dtype=np.float32)

    # Process in batches to avoid memory overload
    for start in range(0, n, batch_size):
        end = min(start + batch_size, n)
        sims = cosine_similarity(vectors[index], vectors[start:end]).flatten()
        sim_scores[start:end] = sims

    # Sort by similarity score (highest first), skip the same internship
    sim_scores[index] = -1  # exclude itself
    top_indices = np.argsort(sim_scores)[::-1][:top_n]

    return new_df.iloc[top_indices][['company_name', 'stipend_min_inr', 'cgpa_cutoff', 'tags']]

# Example: Get top 5 internships similar to the first one
recommendations = recommend_internships(0, top_n=5, batch_size=2000)
recommendations
```

```python
import numpy as np
def recommend_from_skills(user_skills, top_n=5, batch_size=2000):
    # Preprocess user input to match your tags format
    user_skills_processed = user_skills.lower().replace(" ", "")
    
    # Vectorize the user input using the same vectorizer you already fitted
    user_vector = cv.transform([user_skills_processed])
    
    n = vectors.shape[0]
    sim_scores = np.zeros(n, dtype=np.float32)
    
    # Compute similarity in chunks
    for start in range(0, n, batch_size):
        end = min(start + batch_size, n)
        sims = cosine_similarity(user_vector, vectors[start:end]).flatten()
        sim_scores[start:end] = sims
    
    # Get top N most similar internships
    top_indices = np.argsort(sim_scores)[::-1][:top_n]
    
    return new_df.iloc[top_indices][['company_name', 'stipend_min_inr', 'cgpa_cutoff', 'tags']]
```

```python
# Example usage:
user_input = "20778"
recommendations = recommend_from_skills(user_input, top_n=5, batch_size=2000)
recommendations
```
